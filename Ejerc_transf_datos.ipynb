# Datos de ventas de clientes.
### Identificar al cliente con mas compras durante esta semana y recompensarlo
### crear nuevas estrategias para la empresa basados en el insight obtenido.

```python
datos_ventas = pd.read_json('/content/dados_vendas_clientes.json')
datos_ventas.head()
```

```python
datos_ventas_normalizados = pd.json_normalize(datos_ventas['dados_vendas'])
datos_ventas_normalizados.head()
```

```python
columnas = list(datos_ventas_normalizados.columns)
columnas
```

```python
datos_ventas_normalizados = datos_ventas_normalizados.explode(columnas[1:])
datos_ventas_normalizados
```

```python
datos_ventas_normalizados.reset_index(drop=True, inplace=True)
datos_ventas_normalizados
```

```python
datos_ventas_normalizados.info()
```

```python
datos_ventas_normalizados['Valor da compra'] = datos_ventas_normalizados['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',', '.').strip())
datos_ventas_normalizados
```

```python
datos_ventas_normalizados['Valor da compra'] = datos_ventas_normalizados['Valor da compra'].astype(np.float64)
datos_ventas_normalizados.info()
```

```python
datos_ventas_normalizados['Cliente'] = datos_ventas_normalizados['Cliente'].str.lower()
datos_ventas_normalizados
```

```python
datos_ventas_normalizados['Cliente'] = datos_ventas_normalizados['Cliente'].str.replace('[^a-zA-Z\']', ' ', regex=True).str.strip()
datos_ventas_normalizados
```

```python
datos_ventas_normalizados.info()
```

```python
datos_ventas_normalizados['Data de venda'] = pd.to_datetime(datos_ventas_normalizados['Data de venda'])
datos_ventas_normalizados.info()
```

```python
orden_por_fecha = datos_ventas_normalizados.groupby(['Cliente'])[['Cliente', 'Data de venda']]
orden_por_fecha.head()
```

```python
total_de_compras = datos_ventas_normalizados.groupby(['Cliente'])['Valor da compra'].sum()
total_de_compras = pd.DataFrame(total_de_compras)
total_de_compras.head(10)
```

```python
total_de_compras.sort_values('Valor da compra', ascending=False, inplace=True)
total_de_compras.head(10)
```

# Administraci坦n de condominios.
### Analizar el retraso de pago de los inquilinos.

```python
datos_condominios = pd.read_json('/content/dados_locacao_imoveis.json')
datos_condominios.head()
```

```python
datos_condominios_normalizado = pd.json_normalize(datos_condominios['dados_locacao'])
datos_condominios_normalizado.head()
```

```python
columnas2 = list(datos_condominios_normalizado.columns)
columnas2
```

```python
datos_condominios_normalizado
```

```python
datos_condominios_normalizado = datos_condominios_normalizado.explode(columnas2[1:])
datos_condominios_normalizado
```

```python
datos_condominios_normalizado.reset_index(drop=True, inplace=True)
datos_condominios_normalizado
```

```python
datos_condominios_normalizado.info()
```

```python
datos_condominios_normalizado['valor_aluguel'] = datos_condominios_normalizado['valor_aluguel'].apply(lambda x: x.replace('$ ', '').replace('reais', '').replace(',', '.').strip())
datos_condominios_normalizado
```

```python
datos_condominios_normalizado['valor_aluguel'] = datos_condominios_normalizado['valor_aluguel'].astype(np.float64)
datos_condominios_normalizado.info()
```

```python
datos_condominios_normalizado['apartamento'] = datos_condominios_normalizado['apartamento'].str.upper()
datos_condominios_normalizado
```

```python
datos_condominios_normalizado['apartamento'] = datos_condominios_normalizado['apartamento'].str.replace('\([^)]*\)', '', regex=True).str.strip()
datos_condominios_normalizado
```

```python
datos_condominios_normalizado['datas_combinadas_pagamento'] = pd.to_datetime(datos_condominios_normalizado['datas_combinadas_pagamento'], format='%d/%m/%Y')
datos_condominios_normalizado['datas_de_pagamento'] = pd.to_datetime(datos_condominios_normalizado['datas_de_pagamento'], format='%d/%m/%Y')
datos_condominios_normalizado.info()
```

```python
datos_condominios_normalizado['dias_atraso'] = (datos_condominios_normalizado['datas_de_pagamento'] - datos_condominios_normalizado['datas_combinadas_pagamento']).dt.days
datos_condominios_normalizado
```

```python
media_de_atraso = datos_condominios_normalizado.groupby(['apartamento'])['dias_atraso'].mean()
media_de_atraso = pd.DataFrame(media_de_atraso)
media_de_atraso
```

```python
media_de_atraso.sort_values('dias_atraso', ascending=False, inplace=True)
media_de_atraso
```
